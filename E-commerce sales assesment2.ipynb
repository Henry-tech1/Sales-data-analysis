Part 1: DATA PREPROCESSING IN PYTHON 

- Import the amazon.csv dataset into a Pandas DataFrame.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('amazon.csv')
```

- Display the first few rows and understand the data structure.

```python
# Display the first few rows
df.head()
```

```python
# Checking the structure of the data
df.info()
```

- Check for missing values and handle them appropriately.

```python
# Check for missing values
df.isnull().sum()
```

```python
df['rating_count'] = df['rating_count'].fillna(df['rating_count'].mode()[0])  # Mode might return a list, so take the first element
```

```python
df.isnull().sum()
```

2. DATA CLEANING

- Convert price-related columns (discounted_price, actual_price) to
numerical format.

```python
# Remove the ₹ symbol and commas, and convert to numerical format desired
df['discounted_price'] = df['discounted_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
df['actual_price'] = df['actual_price'].replace({'₹': '', ',': ''}, regex=True).astype(float)
```

```python
df['discounted_price']
```

```python
df['actual_price']
```

```python
df.info()
```

- Convert rating_count to integer values (remove commas if present).

```python
#Remove commas and convert to integer
df['rating_count'] = df['rating_count'].replace({',':''}, regex=True).astype(int)
```

```python
df.info()
```

- Extract relevant product categories from category column.

```python
df['category'] = df['category'].str.split(' , ').str[0]
```

3. DATA TRANSFORMATION

- Create a new column for discount_amount (actual_price -discounted_price)

```python
# Calculate discount amount (actual_price - discounted_price)
df['discount_amount'] = df['actual_price'] - df['discounted_price']
```

```python
df.columns
```

- Convert discount_percentage into a numeric value.

```python
# Convert discount_percentage to numeric (remove % symbol and convert to float)
df['discount_percentage'] = df['discount_percentage'].replace({'%': ''}, regex=True).astype(int)
```

- Extract the year of reviews if a timestamp is present.

```python
# Assuming there's a 'timestamp' column in datetime format
if 'timestamp' in df.columns:
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df['review_year'] = df['timestamp'].dt.year # way to access various date and time properties on a datetime column
else:
   print("Timestamp column is missing. Cannot extract year.")
```

```python
cleaned_df = df
```

Part 3: DATA ANALYSIS IN PYTHON 

6. Exploratory Data Analysis (EDA)
- Show the distribution of ratings using a histogram.

```python
# Define the bin edges explicitly to cover the 1 to 5 rating scale
bins = np.arange(0.5 , 5 , 1) 
# Plot the histogram with the specified bins
plt.hist(cleaned_df['rating'], bins=bins, edgecolor='black')

# Add title and labels
plt.title('Distribution of Ratings')
plt.xlabel('rating')
plt.ylabel('frequency')
plt.show()

```

- Find the correlation between discount_percentage and rating.

```python
# Now calculate the correlation
correlation = df[['discount_percentage', 'rating_count']].corr()
print(correlation)
```

- Identify top-selling product categories.

```python
top_categories = df.groupby('category')['rating_count'].sum().sort_values(ascending=False).head(10)
print(top_categories)
```

- Analyze the relationship between discount and rating_count.

```python
plt.scatter(df['discount_percentage'], df['rating_count'])
plt.title('Discount vs Rating Count')
plt.xlabel('discount Percentage')
plt.ylabel('rating Count')
plt.show()
```

- Create a bar chart showing the top 10 highest-rated products.

```python
top_rated = df[['product_name', 'rating_count']].sort_values(by='rating_count', ascending=False).head(10)
top_rated.plot(kind='bar', x='product_name', y='rating_count', title='Top 10 Highest-Rated Products', legend=False)
plt.xticks(rotation=90)
plt.show()
```

- Plot a scatter plot of discount_percentage vs rating.

```python
plt.scatter(df['discount_percentage'], df['rating'])
plt.title('Discount Percentage vs Rating')
plt.xlabel('Discount Percentage')
plt.ylabel('Rating')
plt.show()
```

- Show the top-selling categories using a pie chart.

```python
# Group by 'category' and sum 'rating_count' to get the sales or review count
category_sales = cleaned_df.groupby('category')['rating_count'].sum().sort_values(ascending=False)

# Get the top 10 selling categories
top_10_categories = category_sales.head(10)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_10_categories, labels=top_10_categories.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Top-Selling Product Categories')
plt.axis() 
plt.show()
```

- Use a heatmap to show correlations between numerical columns.

```python
# Identify numeric columns only
numeric_df = cleaned_df.select_dtypes(include=['number'])

# Compute the correlation matrix using only numeric columns
correlation_matrix = numeric_df.corr()

# Create the heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

# Set the title and display the plot
plt.title('Correlation Heatmap')
plt.show()
```
